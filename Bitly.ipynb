Este ejemplo nos permite estudiar los siguientes aspectos:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pathlib as path
```

```python
ruta= path.Path.cwd() / '00 - datasets' / 'datasets' / 'bitly_usagov' / 'example.txt'
```

### Operadores Básicos

np.where()
np.argwhere()
np.select()
np.piecewise()

```python
df=pd.read_json(ruta, lines=True)
df.dtypes
```

```python
df = pd.read_json(ruta, lines=True)
df['a'] = np.where(df.a.str.contains('Windows'), 'Windows', df.a)
df
```

```python
df = pd.read_json(ruta, lines=True)

condlist = np.array([
            df['a'].str.contains('Windows'),
            df['a'].str.contains('Linux'),
            df['a'].str.contains('Macintosh'),
            df['a'].str.contains('Google')], dtype=bool)

choicelist = [
                'Windows',
                'Linux',
                'Macintosh',
                'Google']

df['a'] = np.select(condlist, choicelist, default='Other')
df
```

```python
df = pd.read_json(ruta, lines=True)

condlist = np.array([
            df['a'].str.contains('Windows'),
            df['a'].str.contains('Linux'),
            df['a'].str.contains('Macintosh'),
            df['a'].str.contains('Google')], dtype = bool)

choicelist = [
                lambda x: 'Windows',
                lambda x: x,
                lambda x: 'Linux',
                lambda x: x]

df['a'] = np.piecewise(df, condlist, choicelist)
df
```
