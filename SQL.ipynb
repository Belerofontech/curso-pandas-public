SQL: SELECT *, RANK() OVER (ORDER BY Age DESC) FROM players

```python
players['Rank'] = players['Age'].transform(
    lambda x: x.rank(ascending=False)).astype(int)
players.sort_values(by=['Age'], ascending=False)
```

SQL: SELECT *, RANK() OVER (PARTITION BY Position ORDER BY Age DESC) FROM players

```python
SQL = pd.read_sql(
    'SELECT *, Row_number() OVER (PARTITION BY Position ORDER BY Age DESC) as Rank FROM players ORDER BY position, age desc', con=db)
SQL
```

```python
players['Rank'] = players.groupby(
    ['Position'])['Age'].transform(lambda x: x.rank())
players.sort_values(by=['Position', 'Age'])
```

SQL: SELECT *, DENSE_RANK() OVER (ORDER BY Age DESC) FROM players

```python
SQL = pd.read_sql(
    'SELECT *, DENSE_RANK() OVER (ORDER BY Age DESC) as Rank FROM players', con=db)
SQL
```

```python
players = pd.DataFrame(data=df_list, columns=df_columns)

players['Rank'] = players['Age'].transform(
    lambda x: x.rank(ascending=False, method='dense')).astype(int)
players.sort_values(by=['Age'], ascending=False)
```
