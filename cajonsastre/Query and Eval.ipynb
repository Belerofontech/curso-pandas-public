# Funciones avanzadas

Estas funciones permite un uso eficiente del dataset y la introducción de operaciones dificiles de implementar por otros medios. Fucionan correctamente con datasets grandes.

https://stackoverflow.com/questions/53779986/dynamic-expression-evaluation-in-pandas-using-pd-eval

```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import sidetable
sns.set()
```

```python
diamonds= sns.load_dataset('diamonds')
```

```python
diamonds.shape
```

```python
diamonds.info()
```

```python
pd.eval('Total = log(diamonds.x * diamonds.y * diamonds.z)', target=diamonds)
```

```python
diamonds.eval("Total = x + y + z")
```

```python
diamonds.eval("""
                    Total = x + y + z
                    Patata = depth + depth
    """)
```

```python
diamonds.query("color == 'E' & clarity == 'SI2' & cut == 'Good'")
```

```python
diamonds = (sns.load_dataset('diamonds')
           .query("color == 'E' & clarity == 'SI2' & cut == 'Good'")
           .eval("""
                       Total = x + y + z
                       Patata = table * price"""
                )
           )
diamonds
```
